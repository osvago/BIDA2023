Review of Calculus tools: Functions in one variable $f(x)$

```python
# importing the required modules
import matplotlib.pyplot as plt
import numpy as np
  
```

Visualization of basic functions and its derivatives

```python
# setting the x - coordinates
x = np.arange(-2, 2, 0.1)
# setting the corresponding y - coordinates
y = np.square(x)
  
# plotting the points
plt.plot(x, y)
  
# function to show the plot
plt.show()
```

Define the Sigmoid Function

```python
# setting the x - coordinates
x = np.arange(-5, 5, 0.1)
# setting the corresponding y - coordinates
def sigmoid(x):
  return np.power(x, 2)-3*np.power(x, 3)+0.5*np.power(x,4)


y = sigmoid(x)
  
# plotting the points
plt.plot(x, y)
  
# function to show the plot
plt.show()
```

Find the derivative and plot its graph

```python
D = np.gradient(y)
plt.plot(x,D)
plt.show()
```

Visialization of 2D functions 

```python
# Import libraries
from mpl_toolkits import mplot3d
 
# Creating dataset
x = np.outer(np.linspace(-3, 3, 32), np.ones(32))
y = x.copy().T # transpose
z = np.square(x) - np.square(y) 
 
# Creating figure
fig = plt.figure(figsize =(14, 9))
ax = plt.axes(projection ='3d')
 
# Creating plot
ax.plot_surface(x, y, z)
 
# show plot
plt.show()
```
